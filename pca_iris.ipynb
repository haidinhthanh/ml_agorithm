```python
import numpy as np
from sklearn import datasets, decomposition
from model.pca import PCA
from utils.image_utils import ImageUtils

# load dataset
np.random.seed(5)

iris = datasets.load_iris()
X = iris.data
y = iris.target

# apply pca
n_components = 3
pca = PCA(n_components)
pca.fit(X)
ImageUtils.plot_explained_variance(pca)
```

```python
# X after dim reduction
X_compressed = np.dot(X, pca.components)
print(f"X_compressed.shape {X_compressed.shape}")
# plot
ImageUtils.plot_iris_in_3d(X_compressed, y)
```
